# Data Manipulation: 
#### With a full dataset the data can now be manipulated for analysis and visualization.

1. Separating the **Positive and Negative Patients** to work with the Cancer Positive patients.
2. Separating the **Chernobyl and non-Chernobyl residing Patients** to work with the Cancer Positive subjects from Chernobyl.
3. **Assigning NUMERICAL Values** to non-numeric data (City, CancerTYPE, and Smoker status).

```python
#Separating Patients based on CancerSTATUS
data_pos = []
data_neg = []
lenrow = 20
i=0
j=0
for i in range(lenrow):
    if data['CancerSTATUS'][j] == 'Positive':
        data_pos.append(data[i])
        j=j+1
        i=i+1
    else:
        data_neg.append(data[i])
        j=j+1
        i=i+1
        continue
pos_array = np.array(data_pos)
neg_array = np.array(data_neg)

print(pos_array) #Cancer Positive Patients
#print(pos_array.shape)
#print(neg_array) #Cancer Negative Patients
#print(neg_array.shape)
```

```python
#Separating Patients based on City
pos_chernobyl = []
pos_other = []
lenrow = 11
i=0
j=5
for i in range(lenrow):
    if pos_array[i][j] == 'Chernobyl':
        pos_chernobyl.append(pos_array[i])
        i=i+1
    else:
        pos_other.append(pos_array[i])
        i=i+1
        
chernobyl_pat=np.array(pos_chernobyl)
other_pat=np.array(pos_other)
print(chernobyl_pat) #Chernobyl Positive Patients
#print(other_pat) #Other Cities Positive Patients
```

###### <u>Assigning NUMERICAL Values</u> to non-numeric data

```python
#Numerical Values to City among Positive Patients
import numpy as np
poscity_num=[]
lenrow = 11
i=0
j=5
for i in range (lenrow):
    if pos_array[i][j] == 'Chernobyl':
        poscity_num.append(1)
        i=i+1
    else:
        if pos_array[i][j] == 'Kiev':
            poscity_num.append(2)
            i=i+1
        else:
            if pos_array[i][j] == 'Pripyat':
                poscity_num.append(3)
                i=i+1
            else:
                poscity_num.append(4)
                i=i+1
print(poscity_num)
```

```python
#Numerical Values to CancerTYPE among Positive Patients
import numpy as np
posctype_num=[]
lenrow = 11
i=0
j=8
for i in range (lenrow):
    if pos_array[i][j] == 'Lung':
        posctype_num.append(1)
        i=i+1
    else:
        if pos_array[i][j] == 'Breast':
            posctype_num.append(2)
            i=i+1
        else:
            if pos_array[i][j] == 'Blood':
                posctype_num.append(3)
                i=i+1
            else:
                if pos_array[i][j] == 'Brain':
                    posctype_num.append(4)
                    i=i+1
                else:
                    posctype_num.append(5)
                    i=i+1
print(posctype_num)
```

```python
#Numerical Values to Smokers and Non-Smokers among Positive Patients
#import numpy as np
possmk_num=[]
p_nsmk=0 #Cancer Positive NON-smokers
p_smk=0 #Cancer Positive smokers
lenrow = 11
i=0
j=9
for i in range (lenrow):
    if pos_array[i][j] =="Y":
        possmk_num.append(1)
        p_smk=p_smk+1
        i=i+1
    else:
        possmk_num.append(0)
        p_nsmk=p_nsmk+1
        i=i+1
print(possmk_num)
print('Cancer Positive smokers:', p_smk)
print('Cancer Positive NON smokers:', p_nsmk)
```

1. 3D plot of of Smoking Habit to City of Patients to Types of Cancer to analyze three different sides of data.

<u>*For the Visualization/Analysis sections:*</u>
- Cities Numerically follow this order: Chernobyl (1), Kiev (2), Pripyat (3), and Yaniv (4)
- Cancer Types Numerically follow this order: Lung (1), Breast (2), Blood (3), Brain (4), and None (5)
- Smoker Status: NON-Smokers (0) and Smokers (1)

```python
#3D Plot- All In One
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib notebook
fig=plt.figure()
ax1= fig.add_subplot(111, projection='3d')

x=poscity_num
y=possmk_num
z=posctype_num


ax1.scatter(x,y,z, c="r", marker="s")
ax1.set_xlabel("City")
ax1.set_ylabel("Smoker")
ax1.set_zlabel("CancerTypes")
plt.show()
```

***Analysis:*** From this 3D scatter plot we have three different visualizations comparing three different data.
- Firstly, we can see Smoker status against Cancer Types and notice that most of these Cancer positive patients are not smokers. This could be indication of the Nuclear Disaster having an effect in the development of cancer within these patients.
- Secondly, we can see the Smoker status against City in which we can see among the positive patients most are non smokers and in relation to the project goals; Chernobyl has both non smokers and smokers with cancer. This could mean that the nuclear disaster might have had something to do with the cancer presence.
- Lastly, we can see the Types of Cancer against different Cities. Looking at this side of the plot we can see the type of cancer that exists within each city. For the purposes of the project goals it seems that patients from Chernobyl are also positive for Lung cancer which might indicate a correlation to the incident.

All though these 3D scatter plots are very informative they are not specific enough. They give us certain glimpses of potential evidence that the Nuclear Disaster had any major effects on the people of Chernobyl and their cancer diagnosis.

*If 3D plot is difficult to understand please scroll down to the bottom for a 2D rendition of the same data.

```python
#Separating Positive Patients by INDIVIDUAL Cities for the Bar Graph
c_count=0
k_count=0
p_count=0
y_count=0
lenrow = 11
i=0
j=0
for i in range (lenrow):
    if pos_array["City"][j] == 'Chernobyl':
        c_count=c_count+1
        j=j+1
    else:
        if pos_array["City"][j] == 'Kiev':
            k_count=k_count+1
            j=j+1
        else:
            if pos_array["City"][j] == 'Pripyat':
                p_count=p_count+1
                j=j+1
            else:
                y_count=y_count+1
                j=j+1
city_count=[c_count, k_count, p_count, y_count]
print(city_count)
```

```python
#Bar Graph
import matplotlib.pyplot as pt
fig, ax = plt.subplots()
#ax=fig.add_axes([0,0,1,1])

cities=["Chernobyl", "Kiev", "Pripyat", "Yaniv"]
ax.bar(cities, city_count, color="b")
ax.set_xlabel('Cities')
ax.set_ylabel('Positive Cancer Count')
pt.show()
```

***Analysis:*** From this BAR Graph we can see the spread of positive cancer patient counts between the 4 cities.
- It is clear that Chernobyl has the highest count for positive cancer patients. This is a potential indicator to the nuclear disaster having a chronic effect on the population of Chernobyl.

Although this bar graph sends a clear message it is still not enough to correlate the nuclear event with the high cancer numbers from the people of Chrenobyl within all of the patients.

```python
#Separating the Cancer-Positive Patients from Chernobyl by Smokers vs NON-Smokers & Lung Cancer vs Other Cancer Diagnosis.

cher_smk=0 #Chernobyl Smokers
cher_nsmk=0 #Chernobyl Non-Smokers
cher_lung=0 #Chernobyl Lung Cancer Patients
cher_nlung=0 #Chernobyl Other Cancer Patients
lenrow = 7
i=0
j=0
for i in range (lenrow):
    if chernobyl_pat["Smoker"][j] == "Y":
        cher_smk=cher_smk+1
        j=j+1
    else:
        cher_nsmk=cher_nsmk+1
        j=j+1
m=0
n=0
for m in range (lenrow):
    if chernobyl_pat["CancerTYPE"][n] == "Lung":
        cher_lung=cher_lung+1
        n=n+1
    else:
        cher_nlung=cher_nlung+1
        n=n+1
print("Chernobyl Smokers:", cher_smk, "Non-Smokers:", cher_nsmk)
print("Chernobyl Lung Cancer Patients:" ,cher_lung, "Non-Lung Cancer", cher_nlung)
```

```python
#Pie Chart 1- Chernobyl Smokers to Non-Smokers
import matplotlib.pyplot as pt
fig1 = pt.figure()
ax1=fig1.add_axes ([0,0,1,1])
chernobyl_smklegend=["Chernobyl Smokers", "Chernobyl NON-Smokers"]
cher_smk_results=[cher_smk, cher_nsmk]
ax1.pie(cher_smk_results, labels=chernobyl_smklegend, autopct="%1.1f%%")
pt.title("Chernobyl Smokers vs NON Smokers")
pt.show()

#Pie Chart 2- Chernobyl Lung Cancer to Non-Lung Cancer
import matplotlib.pyplot as pt
fig2 = pt.figure()
ax2=fig2.add_axes ([0,0,1,1])
chernobyl_smklegend=["Chernobyl Lung Cancer", "Chernobyl Other Cancer"]
cher_smk_results=[cher_lung, cher_nlung]
ax2.pie(cher_smk_results, labels=chernobyl_smklegend, autopct="%1.1f%%")
pt.title("Chernobyl Lung Cancer vs Other Cancer Types")
pt.show()
```

***Analysis:*** From these PIE CHARTS we can see the ratio of Chernobyl Smokers to Non-Smokers AND Chernobyl Lung Cancer to Other Cancer patients.

- Pie Chart 1, shows that nearly 86% of the Chernobyl Cancer-Positive patients are _NON-SMOKERS._ This means more NON-SMOKERS from Chernobyl have Cancer.
- Pie Chart 2, shows that nearly 86% of the Chernobyl Cancer-Positive patients have _LUNG-CANCER_ This means that among these patients the presence of LUNG-CANCER is higher.

**Putting this together shows that *MOST* of the Cancer-Positive patients of Chernobyl are Non-Smokers with Lung Cancer. 
This points directly to the aftermath effects of the Nuclear Disaster on the Chernobyl population as the consequences of this event has had seriously harmful effects on the people.**
- This fulfills our goal in connecting the Nuclear-event with the high presence of Lung Cancer within the people of Chernobyl.

```python
import matplotlib.pyplot as pt
age_pos=[chernobyl_pat['Age'], other_pat['Age']]
group_labels = ['Chernobyl Positive Patients', 'Other Positive Patients']
fig, ax = plt.subplots()

b=ax.boxplot(age_pos)
ax.set_xticklabels(group_labels)
pt.show()
```

***Analysis:*** From this BOX-PLOT we can see the AGE range between the cancer-positive patients of Chrenobyl compared with the cancer-positive patients of the other cities.

The purpose of using the AGE factor is to see how the timeline from the event (36 Years ago) to the time of this study (2022) and how the people of Chernobyl compare.

- From the looks of this box plot it seems that the average AGE of the Positive-Cancer patients of Chernobyl is 40 years old. 

This makes sense as they would have been affected by the Nuclear-Event at a young age and developed the Cancer later on (as Non-smokers aswell).
